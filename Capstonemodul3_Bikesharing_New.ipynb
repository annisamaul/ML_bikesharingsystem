### **Algorithm & Metric Evaluation**

**Machine Learning algorithm** that will be used **(potentially suitable)** to predict the number of bicycles to rent:

1. **Linear Regression**: This algorithm is suitable for regression problems where we are trying to predict a continuous value, such as the number of bicycles rented. This is one of the simplest and easiest to interpret regression algorithms.

2. **Decision Tree Regression**: Decision tree regression is suitable for problems that involve modeling non-linear data structures. This makes it possible to build more complex models and capture non-linear relationships between features and target variables.

3. **Random Forest Regression**: Random forest is a development of decision trees that combines several decision trees to increase accuracy and reduce overfitting. It is often used for complex regression problems.

4. **Gradient Boosting Regression**: Gradient boosting regression is an ensemble learning technique that builds a strong predictive model by combining multiple weak models, typically decision trees. It sequentially adds predictors to the ensemble, each one correcting its predecessor, resulting in a powerful predictive model.

5. **Support Vector Regression (SVR)**: SVR is suitable for regression tasks when dealing with smaller datasets. It works well in high-dimensional spaces and is effective when the number of features is greater than the number of samples. SVR aims to find the optimal hyperplane that maximizes the margin, allowing for better generalization to unseen data.

6. **K-Nearest Neighbors Regressor** : will predict the number of bicycles to be rented based on the k nearest neighbors of the point to be predicted. Suitable for use if patterns in the data can be explained by nearest neighbors.

7. **Lasso and Ridge** : Lasso and Ridge Regressions are suitable when there are many features in the data and it is necessary to control overfitting. They can help in finding the most important features and dealing with multicollinearity.

8. **XGBoost Regressor and AdaBoost Regresso**r: XGBoost and AdaBoost are powerful ensemble models that are suitable for use when there is high complexity in the data and high performance in predictions is required. They will capture complex patterns in the data and can provide accurate predictions in a variety of scenarios.


**Mean Absolute Percentage Error (MAPE)** was chosen as the **evaluation metric** for the following reasons:

- MAPE provides the average error in percentage, which is easy to understand and interpret. This allows us to know how accurate the model is in predicting the number of bikes rented in terms of a percentage of the actual value.
- MAPE is not affected by extreme values or outliers. This allows us to understand the overall prediction error without being unduly influenced by unusual or unrepresentative data.

For analysis purposes, we create a new column to define whether the date is a weekend or weekday.

```python
import datetime

# Determine whether a date is a weekend or weekday
def is_weekend(date_string):
    date_obj = datetime.datetime.strptime(date_string, "%Y-%m-%d")
    if date_obj.weekday() >= 5:
        return 1
    else:
        return 0

# Make a new column
bike.insert(2, 'is_weekend', bike['dteday'].apply(is_weekend))

# Display dataset
bike.head()
```

### **`Feature Engineering`**

Feature engineering is the process of creating or changing features in a dataset into a representation that is more informative or more suitable for use in modeling. The main goal of feature engineering is to improve the performance of predictive models by introducing more relevant features, reducing feature dimensions, overcoming multicollinearity problems, and dealing with other problems that may exist in the dataset.

```python
#Create pipeline for preprocessing
numeric_transformer = Pipeline(steps=[
    ('scaler', RobustScaler())  #lebih cocok ke data yang banyak outliernya, tidak berdistribusi normal
])

categorical_features = ['weathersit','season']
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore',drop='first'))])


# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],remainder='passthrough')

preprocessor
```

1. **numeric_transformer:**

- **RobustScaler()** is used for scaling numerical features. RobustScaler uses a scaling method based on median and interquartile range (IQR), so it is more resistant to outliers than standard scaling methods. This is useful because we have decided to retain outliers in our dataset.

2. **categorical_transformer**:

- **OneHotEncoder(drop='first')** is used to encode categorical features into binary vectors. OneHotEncoder generates a binary vector that has a value of 0 or 1, where each value represents the presence or absence of a category in that feature. The drop='first' parameter removes the first category generated by OneHotEncoder, thus avoiding multicollinearity in our model.

3. **preprocessor:**

- **ColumnTransformer** is used to combine preprocessing steps for numerical and categorical features. This allows us to apply different transformations to different features simultaneously. In this case, we have two transformers: one for numerical features and one for categorical features.

This helps improve target distribution and variability and improves prediction quality.

> **Model Performance Evaluation**

This stage is an **evaluation of the performance of several regression models** whose targets have been transformed using the evaluation metrics MAPE (Mean Absolute Percentage Error), MAE (Mean Absolute Error), and RMSE (Root Mean Squared Error) through cross-validation with 5 folds. . This stage involves iterating through various target-transformed regression models and building a pipeline that includes preprocessing and the model. Each model was evaluated using the metrics mentioned above, and the mean value and standard deviation of each metric were calculated. The results of this evaluation are then combined into a DataFrame for further analysis.

```python
scores = []
mean = []
std = []

# Modeling is carried out on a logarithmic scale, but then inverted again for interpretation
log_linreg = TransformedTargetRegressor(linreg, func=np.log, inverse_func=np.exp)
log_gboost = TransformedTargetRegressor(gboost, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_svr = TransformedTargetRegressor(svr, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_lasso = TransformedTargetRegressor(lasso, func=np.log, inverse_func=np.exp)
log_ridge = TransformedTargetRegressor(ridge, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)
log_adb = TransformedTargetRegressor(adb, func=np.log, inverse_func=np.exp)

# Candidate algorithms
models = [log_linreg,log_gboost,log_rf,log_svr,log_dt,log_knn,log_lasso,log_ridge,log_xgb,log_adb]

for model in models:
    pipe = Pipeline([
        ('preprocess', preprocessor),
        ('model', model)
    ])
    
    # Perform cross-validation for MAPE
    cv_scores_mape = cross_val_score(pipe, X_train, y_train, cv=5, scoring='neg_mean_absolute_percentage_error')
    
    # Perform cross-validation for MAE
    cv_scores_mae = cross_val_score(pipe, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')
    
    # Perform cross-validation for RMSE
    cv_scores_rmse = cross_val_score(pipe, X_train, y_train, cv=5, scoring='neg_mean_squared_error')

    # Append scores to the list
    scores.append((cv_scores_mape, cv_scores_mae, cv_scores_rmse))
    
    # Calculate mean and standard deviation of scores
    mean.append((cv_scores_mape.mean(), cv_scores_mae.mean(), cv_scores_rmse.mean()))
    std.append((cv_scores_mape.std(), cv_scores_mae.std(), cv_scores_rmse.std()))
```

### **`Hyperparameter Tuning`**

**Hyperparameter tuning** is the process of finding the optimal combination of values for the model hyperparameters used in machine learning algorithms. Hyperparameters are parameters that are not learned directly from data during the model training process, but must be determined before the training process begins. The goal of hyperparameter tuning is to improve model performance by finding the combination of hyperparameters that produces the best results, such as accuracy or lower error, based on the selected evaluation metrics. This process can be carried out using the **grid search method, random search**, or other optimization techniques.

1. **Grid Search**: In grid search, we define a list of possible values for each hyperparameter we want to tune. Grid search then tries all possible combinations of hyperparameter values and evaluates the model for each combination. With this approach, grid search systematically explores the entire hyperparameter space.

2. **Random Search**: In contrast, in random search, we randomly select values for each hyperparameter in each iteration. This means that each iteration can produce a different combination of hyperparameters. Random search does not explore the entire hyperparameter space like grid search, but is often more efficient because it can find good solutions with fewer iterations.
